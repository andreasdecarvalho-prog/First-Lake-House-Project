# Create dict to automate file ingestion -->

```python
from pathlib import Path

folder_path = Path("/Volumes/workspace/bronze/source_systems")

# Dictionary: {filename: filepath}
file_dict = {f.name: str(f) for f in folder_path.rglob("*") if f.is_file()}

print(file_dict)
```

# Loop through the list and ingest the data into bronze layer

```python
for name, path in file_dict.items():
  df = (
      spark.read.option("header", "True")
      
      .option("inferSchema", "True")
      .csv(path)
  )
  
  df.write.mode("overwrite").saveAsTable(f"workspace.bronze.{name.split('.')[0]}")
```
