# Creates Gold layer tables

## Create dimensional "customers" table

```python
%sql
CREATE TABLE workspace.gold.customers AS
SELECT id, key, 
  firstname || ' ' || lastname AS full_name,
  gender,
  marital_status
FROM workspace.silver.cust_info 
```

## Create dimensional "products" table

```python
%sql
CREATE TABLE workspace.gold.products AS 
SELECT prd_id AS id, prd_key AS key, 
    prd_name AS name, prd_cost AS cost,
    category, subcategory 
FROM workspace.silver.prd_info AS t1
JOIN workspace.silver.px_cat_g1v2 AS t2
    ON LEFT(t1.prd_key, 5) = t2.id
```

## Create Fact table "sales"

```python
%sql
CREATE TABLE workspace.gold.sales AS
SELECT order_id, product_id, customer_id, 
    order_date, ship_date, due_date,
    price
FROM workspace.silver.sales_details
```
