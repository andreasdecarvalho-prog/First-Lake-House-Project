```python
# basic init with impor
from pyspark.sql.functions import col, trim, try_to_date
```

```python
# basic df initialization and parsing

def df_init(file_path):
    df = spark.read.table(file_path)
    df = clean_whitespaces(df)
    df = handle_nulls(df)

    table_name = file_path.split("workspace.bronze.")[1]
    return df, table_name
```

```python
# gets rid of whitespace in rows

def clean_whitespaces(df):
    return df.select(*(trim(df[col]).alias(col) for col in df.columns))
```

```python
# changes values like "M" to "male"

def replace_gender(df, column):
    return df.replace({"M": "male", "F": "female"}, subset=[column])
```

```python
# handles null values

def handle_nulls(df):
    df = df.fillna("unknown")
    df = df.replace(' ', "unknown")
    return df
```

```python
# writes the parsed data into the silver table
def writing_silver_table(df, table_name):
    (
    df.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable(f"workspace.silver.{table_name}")
)
```
