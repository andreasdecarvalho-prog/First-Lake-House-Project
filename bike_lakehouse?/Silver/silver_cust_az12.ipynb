## Init

```python
from pyspark.sql.functions import col
```

## Read from bronze table

```python
df = spark.read.table("workspace.bronze.cust_az12")
```

# Transform and normalize data

## Replace NULLS

```python
df = df.fillna({"GEN": "unknown"})
df = df.replace(' ', "unknown")
```

## Replace single char values for better readability

```python
df = df.replace(['M', 'F'], ['Male', 'Female'])
```

## Rename columns for better readability

```python
df = df.withColumnsRenamed({"CID": "customer_id", "BDATE": "birth_date", "GEN": "gender"})
```
